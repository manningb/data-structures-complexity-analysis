# Resource Allocation Problem
> disclaimer. this is an inspiration drew from an actual course project in computer science programme. similarity is being mindfully minimised, however due to time constraints, actual dummy data were copy-pasted from the project, even though they were also synthesized. 

In some teaching programmes here in the computer science school, we run a so-called final year/semester project: students get to pick topics from a pool, in preferences in a descendent order, and they (the topics) are alotted to the student based on the preferences and lottery. The lottery part is in place because there might be some _hot_ topics which get a lot of top preferences from multiple students, and each project should only be done by 1 student. There are more project topics in the pool than the number of students, not to mention, that some students even come up with their own ideas to be approved as project, so they don't consume any from the pool.

The rules of preference settings are
  1. each student can name up to 10 project topics for her/his preferences
  2. self-promoted and approved project count as alloted, no preference list is then necessary, or even be considered.

The rules of resource allocation are
  1. near optimality on allotting preferences based on rank/order
  2. on contested topics $\rightarrow$ student assignments, lottery is played
  3. on unfulfilled student assignments, lottery is played from the rest of pool
  
Further constraints on the problem (or the data, more precisely)
  1. project topics are presented in project names, and they are unique, consistent. I.e. no same topic be described in 2 different names.
  2. student names are unique

### [Methodology] How to exploit these properties, for example:
One thing is sure, regardless what solution we might come up with, there bound to be a lot of string comparisons in the solution. The fact that the names are unique helped us a lot, but we still need to determine whether 2 strings are the same. We need to make the comparison as efficient as possible to make higher-order algorithms not having surprises in performance evaluations. 

**There are many ways to do things efficiently (or at least look smart), the key is to make sound argument for your decisions.**

In practice, problems are often way to big for you to make right decision early on. Even for toy problem like this one, there are still surprises in the performance where you won't find explanations from a textbook that easily. For example, say I decide between using hashing and prefix tree (trie) to do string comparison. I know both their strengths and weaknesses and can recite their performance number from memory. But the fact I don't know what the full extend of my input data is, I can't say the hashing is going to out perform trie or vice versa. 

In this case, I implement both and test them both at the end with a simple swap (this is also often called A/B test). This way, my argument for a final decision has scientific merit: I did the experiment and data speaks volume.

### Preview: Design and Strategy for solving the problem

So, the revelation that we don't need actively comparing strings during the bulk calculation is that the project topics become static once we initiate the (allocation) algorithm. 

    --load all the rows--> |OriginalData| --build a set of projects--> |ProjectSet|
    
    |ProjectSet|
        {key : int, value: string} // value is project name

So, we can compare them at the beginning only once, and transform the entire dataset by replacing all the project names with just unique `project_key`. 

Moving on, we should get the next most obvious trivialities out of the way: class for students, projects, and maybe assignments.

    |Student|
        +attr name : string
        -attr project_list : []int
        +func add_project(project) : nil
        +func get_project_list() : []int
        +func set_preassigned_project(project) : nil
        +func has_preassigned_project() : bool

    
    |Assignment|
        -attr student : Student
        -attr project : int
        +func score() : int // return a score signalling how satisfied this student is with the assignment


By now we have stumble upon a gold mine: `score` function! The solution will reach *near optimality* if we can minimise the students' dissatisfactions. The most intuitive and seemingly not smart way to solve it is try to guess some assignment and evaluate the result, if the satisfaction is lowered, we undo the guess and try again. So we modify the original design with this:

    |Student|
        +attr name : string
        -attr project_list : []int
        +func add_project(project) : nil
        +func get_project_list() : []int
        +func set_preassigned_project(project) : nil
        +func has_preassigned_project() : bool
        +func random_project() : Project // pick random from preference list

    
    |Assignment|
        -attr student : Student
        -attr project : int
        -attr last_assigned : Project
        +func random_assign() : nil // remember last assigned and pick random using student's random_project()
        +func undo_assign() : nil // revert back to last assigned
        +func score() : int // return a score signalling how satisfied this student is with the assignment

    

With this skeleton thoughts been put down on paper, we can start the introduction of the main solution design. Ideally, you should provide several design for the main solver to be evaluated against each other, providing performance comparisons and aspects like limitation vs generic/robustness. For now I only have time to code brute force and simulated annealing method. It is not hard to rewrite and expand to adopt methods like genetic algorithms for this task. 

The brute force method basically guess randomly for all assignment and calculate the cumulated satisfaction score, we run this enough of times and pick the best score and print out their assignments. While the name brute force implied naivety and imperfection, this particular implementation is an approximate of simulated annealing method. I have also written a textbook version of simulated annealing, but it did not out-perform the brute force method, partly also due to limitation by action to obtain new configuration: due to relunctance of writing more code, I basically used whatever is originally allowed in the Assignment interface, the action results in a very gentle modification to the system configuration; also my simulated annealing implementation could theorectically produce worst result because there is no penalty or limitation to duplicated assignment of projects.

## Implementation and Testings

Let's start by implementing utility methods, getting the easiest out of the way:

```python
import random
from typing import List


class RNG:
    def __init__(self, seed=124):
        self._seed = seed
        random.seed(self._seed)
        self._state = random.getstate()

    def reset(self):
        random.setstate(self._state)

    def pick_from(self, lst: set) -> int:
        if not lst:
            raise ValueError("lst should not be empty")
        return tuple(lst)[random.randrange(0, len(lst))]
```

#### The classes for main entities

```python
class Student:
    """i'm lazy, but you should write meaningful comment here"""

    def __init__(self, student_name: str):
        self.student_name = student_name
        self.preassigned = -1
        self._preferences = []

    def add_project(self, project: int):
        self._preferences.append(project)

    def set_preassigned(self, project: int):
        self.preassigned = project

    def has_preassigned(self) -> bool:
        return self.preassigned != -1

    def get_preferences(self) -> List[int]:
        return self._preferences

    def randomly_pick(self, rng: RNG, available: set) -> int:
        if self.has_preassigned():
            return self.preassigned

        intersection = available.intersection(set(self.get_preferences()))
        if intersection:  # intersection is not empty
            return rng.pick_from(intersection)
        else:
            return rng.pick_from(available)
```

```python
class DataSet:
    """i'm lazy, but you should write meaningful comment here"""

    def __init__(self, file_name):
        import csv
        with open(file_name, 'r') as tsv:
            reader = csv.DictReader(tsv, dialect="excel-tab")
            row_dict_list = [row for row in reader]

        pref_labels = ["Preference " + str(i) for i in range(1, 11)]

        # add all project name strings to the list, if it's not empty
        all_project_names = [
            row[pref] for row in row_dict_list
            for pref in pref_labels if row[pref]
        ]

        # implied comparison and make a set out of it
        self._set = set(all_project_names)
        ids = [i for i in range(len(self._set))]

        # make a temporary transform map e.g. {'name': 1234} to construct
        # student object
        transform_map = dict(zip(self._set, ids))

        # this is the map we want to retrieve project name
        self._map = {pid: name for name, pid in transform_map.items()}
        self._general_available = set(self._map.keys())
        self._students = []
        for i in range(len(row_dict_list)):
            row = row_dict_list[i]
            s = Student(row['Student Name'])
            if row['Prearranged'] != "Yes":
                for pref in pref_labels:
                    if row[pref] != '':
                        s.add_project(transform_map[row[pref]])
            else:
                # if 'Yes', the first preference is the pre-arranged
                exclusive_title = row[pref_labels[0]]
                exclusive_id = transform_map[exclusive_title]
                s.set_preassigned(exclusive_id)
                self._general_available.remove(exclusive_id)

            self._students.append(s)

    def get_project_name(self, pid) -> str:
        return self._map[pid]

    def get_project_pid_set(self) -> set:
        return self._general_available

    def __getitem__(self, index):
        return self._students[index]

    def __len__(self):
        return len(self._students)
```

```python
class Assignment:
    """i'm lazy, but you should write meaningful comment here"""

    def __init__(self, student: Student):
        self.student = student
        self.assigned = None
        self._last_assigned = None

    def assign(self, project):
        self.assigned = project

    def random_assign(self, rng: RNG, available: set):
        self._last_assigned = self.assigned
        self.assigned = self.student.randomly_pick(rng, available)

    def commit(self, available: set):
        if self.student.has_preassigned() or not available:
            return
        available.remove(self.assigned)

    def undo(self, available: set):
        self.assigned = self._last_assigned
        if self.student.has_preassigned() or not available:
            return
        available.add(self.assigned)

    def score(self):
        """
        score is the satisfaction value of a student towards an assignment.
          - if the student has pre-assigned project, the score will always be
           10.
          - if the student gets an assignment from her/his pref-list, then the
           score will be reflected according to the project order in the list.
            value from 1 to 10
          - if the student gets an assignment from outside the pre-list, then
           the satisfaction is 0
        """
        if self.student.has_preassigned():
            return 10
        try:
            index = self.student.get_preferences().index(self.assigned)
            return 10 - index
        except ValueError:
            return 0

    def __str__(self):
        return "Project ID: {} -> Student: {}".format(
            self.assigned, self.student.student_name
        )
```

```python
class Trial:
    def __init__(self, dataset: DataSet, rng: RNG):
        self._dataset = dataset
        self.trial_project_pool = dataset.get_project_pid_set().copy()
        self._score = 0
        self._rng = rng
        self._assignments = []
        for s in dataset:
            self._assignments.append(Assignment(s))

        self.run()

    def run(self):
        for a in self._assignments:
            a.random_assign(self._rng, self.trial_project_pool)
            a.commit(self.trial_project_pool)
            self._score += a.score()

    def reset(self, return_set: set):
        self._score = 0
        for a in self._assignments:
            a.undo(return_set)

    def score(self):
        return sum([a.score() for a in self.get_assignments()])

    def simulated_anealing(self, bound=5000):
        import math
        configuration = set(self._assignments.copy())
        touchable = [
            a for a in configuration if not a.student.has_preassigned()
        ]
        preassigned: set = configuration.difference(touchable)

        def get_energy(lst):
            return sum([a.score() for a in lst])

        THRESHOLD = random.random()

        def step(lst: List[Assignment]):
            pair = random.sample(lst, 2)
            project_0 = pair[0].assigned
            project_1 = pair[1].assigned
            pair[0].random_assign(self._rng, set([project_1]))
            pair[1].random_assign(self._rng, set([project_0]))
            for i in range(len(lst)):
                if lst[i].student == pair[0].student:
                    lst[i] = pair[0]
                if lst[i].student == pair[1].student:
                    lst[i] = pair[1]
            return lst

        def stride(lst, width=4):
            for i in range(width):
                lst = step(lst)
            return lst

        pool = touchable
        for temperature in range(bound, 1, -1):
            energy = get_energy(pool)
            move = pool.copy()
            new_pool = stride(move)
            new_energy = get_energy(new_pool)
            delta = new_energy - energy
            if delta > 0:
                pool = new_pool
            elif math.pow(math.e, (delta / float(temperature) )) > THRESHOLD:
                pool = new_pool

        self._assignments = list(preassigned.union(pool))

    def __lt__(self, other):
        return self.score() < other.score()

    def __gt__(self, other):
        return self.score() > other.score()

    def __eq__(self, other):
        return self.score() == other.score()

    def __ne__(self, other):
        return self.score() != other.score()

    def get_assignments(self):
        return self._assignments

    def __str__(self):
        assignments = self.get_assignments()

        def describe(a):
            return "{} | Project name: {}".format(
                a, self._dataset.get_project_name(a.assigned)
            )

        return "Trial score: {}\nwith the following assignments:\n{}".format(
            self.score(), "\n".join([describe(a) for a in assignments])
        )
```

#### Bruteforce Trial run and result

```python
dataset = DataSet('input.tsv')

best = None
rng = RNG(42)
num = 5000
for i in range(num):  # 100 epochs
    t = Trial(dataset, rng)
    # print("Trial {}'s score: {}".format(i, t.score()))
    if best is not None:
        if t > best:
            best = t
    else:
        best = t

print("Ran {} trials, the best trial is {}".format(num, best))
```

#### Comparing the run time for both methods

```python
d = DataSet('input.tsv')
rng = RNG(42)

%matplotlib inline
from matplotlib import pyplot as plt

def bruteforce(num, dataset, rng):
    best = None
    for i in range(num):  # 100 epochs
        t = Trial(dataset, rng)
        # print("Trial {}'s score: {}".format(i, t.score()))
        if best is not None:
            if t > best:
                best = t
        else:
            best = t
    print("bruteforce on {}, gets {} ".format(num, best.score()))
    return best.score()


def simann(bound, dataset, rng):
    t = Trial(dataset, rng)
    t.simulated_anealing(bound)
    print("simulated annealing with bound {}, gets {} ".format(bound, t.score()))
    return t.score()


times = [1, 10, 100, 1000, 10000]


```

```python
%time scores = [bruteforce(i, d, rng) for i in times]
plt.plot(times, scores)
```
