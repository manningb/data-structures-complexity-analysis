```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

To test n bottles at we need to have testers drink only from n-1 bottles.

We require the ceiling of $log_{2}((n-1)+1)=log_{2}(n)$ bits to represent the numbers from 1 to n-1 in binary.
Hence the maximum number of deaths will be if a bottle with maximum number of 1's in its binary representation is the one poisoned i.e. The ceiling of $log_{2}(n)$.


```python
fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')

n = np.arange(3, 1000)

fn = np.ceil(np.log2(n))

for c in np.arange(1.01, 1.5, 0.05):
    boundary = c*np.log2(n)
    plt.plot(n, boundary, '--', linewidth=.5, label="c"+str(c))
    
# for m in n:
#     fn = math.ceil(math.log(m,2))
#     plt.plot(m, fn)

plt.plot(n,fn, linewidth=3, label="f(n)")
plt.grid()
plt.xlabel('n')
plt.ylabel('fn')
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.show()
```

So for large enough n, c = 1.16 gives us a bound on the growth of testers as n grows.
i.e. $$ log_{2}(n) <= c * log_{2}(n), n >= 150 $$
